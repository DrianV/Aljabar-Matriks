Nomor 2 (Tensorflow)

```python
import tensorflow as tf

#Lakukan Tahapan Pada Soal 1 Dengan Menggunakan Tensorflow

#Buat Variabel Skalar x Dan Simpan Suatu Nilai Skalar
x_tf = tf.Variable(78)
print(x_tf)
```

```python
#Tampilkan Tipe Data Dari x_tf
print(f'Tipe Data Variabel x_tf Adalah: {type(x_tf)}')
```

```python
#Buat Variabel Skalar y_tf Dan Simpan Suatu Nilai Skalar
y_tf = tf.Variable(79)
print(y_tf)
```

```python
#Tampilkan Tipe Data Dari y_tf
print(f'Tipe Data Variabel y_tf Adalah: {type(y_tf)}')
```

```python
#Jumlahkan x_tf Dan y_tf
tf_sum = x_tf + y_tf
print(tf_sum)
```

```python
#Buat Variabel Bernama x_float_tf Dan Simpan Suatu Nilai Desimal
x_float_tf = tf.Variable(3.14)
print(x_float_tf)
```

```python
#Jumlahkan x_float_sum Dengan y_tf
y_float_tf = float(y_tf)
x_float_sum = x_float_tf + y_float_tf
print(x_float_sum)
```

```python
#tampilkan tipe data dari variabel x_float_tf
print(f'Type dari x_float_tf adalah: {type(x_float_tf)}')
```
