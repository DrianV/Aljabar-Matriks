## Nomor 3


### Pytorch

```python
import torch as pt

# buat vektor x_pt dengan ukuran 3x1 menggunakan pytorch dan ukuran dari vektor x_pt
x_pt = pt.tensor([13, 17, 30])
print(x_pt)
x_pt.shape
```

```python
# buat vektor y_pt dengan ukuran 3x1 menggunakan pytorch dan ukuran dari vektor y_pt
y_pt = pt.tensor([18, 22, 40])
print(y_pt)
y_pt.shape
```

```python
# lakukan transpose pada vektor x_pt dan simpan pada variabel x_pT
x_pT = x_pt.T
print(x_pT)
```

### Tensorflow

```python
import tensorflow as tf

# buat vektor x_pt dengan ukuran 3x1 menggunakan tensorflow dan ukuran dari vektor x_pt
x_tf = tf.Variable([13, 17, 30], dtype = tf.int32)
print(x_tf)
x_tf.shape
```

```python
# buat vektor y_pt dengan ukuran 3x1 menggunakan tensorflow dan ukuran dari vektor y_pt
y_tf = tf.Variable([18, 22, 40], dtype = tf.int32)
print(y_tf)
y_tf.shape
```

```python
# lakukan transpose pada vektor x_tf dan simpan pada variabel x_tF
x_tF = tf.transpose(x_tf)
print(x_tF)
x_tF.shape
```
