## Nomor 1

```python
import numpy as np

# buat vektor x dengan ukuran 3x1
x = np.array([[10], [11], [12]])
print(x)
```

```python
# hitung nilai L2 norm dari vektor x
(10**2 + 11**2 + 12**2)**0.5
```

```python
norm_x = np.linalg.norm(x)
print(norm_x)
```

## Nomor 2

```python
# buat vektor v dengan elemen (3,4)
v = np.array([[3], [4]])
print(v)
```

```python
# hitung nilai L2 norm dari vektor v
(3**2 + 4**2)**0.5
```

```python
# tampilkan unit vector dari vektor v
norm_v = np.linalg.norm(v)
print(norm_v)
```

```python
# hitung panjang unit vektor dari vektor v
unit_v = v / np.linalg.norm(v)
np.linalg.norm(unit_v)
```
