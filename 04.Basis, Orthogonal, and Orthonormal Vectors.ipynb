## Nomor 1

```python
import numpy as np

# buat vektor i dengan elemen (1,0) dan vektor j dengan elemen (0,1)
i = np.array([1, 0])
j = np.array([0, 1])
print(i)
print(j)
```

```python
# buktikan bahwa vektor i dan vektor j itu orthogonal
dot_product = np.dot(i, j)
print(f"Hasil dot product i dan j : {dot_product}, jika hasilnya 0 maka i dan j orthogonal, jika tidak maka i dan j tidak orthogonal.")

if dot_product == 0:
    print("i dan j adalah orthogonal")
else:
    print("i dan j bukan orthogonal")
```

## Nomor 2

```python
# Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
w = np.array([2, -1])
dot_product_vw = np.dot(v, w)
print(f"Hasil dot product v dan w : {dot_product_vw}, jika hasilnya 0 maka v dan w orthogonal, jika tidak maka v dan w tidak orthogonal.")

if dot_product_vw == 0:
    print("v dan w adalah orthogonal")
else:
    print("v dan w bukan orthogonal")
```
